```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
def rename_columns(c, year):
    columns=['COUNTRY OF NATIONALITY']
    for x in c:
        if x !="COUNTRY OF NATIONALITY":
            columns.append(f"{year} {x}")
    return columns
```

```python
# Take in all of our traveler data and read it into pandas
travel_2016 = "../Resources/2016_travelers.csv"
travel_2017 = "../Resources/2017_travelers.csv"
travel_2018 = "../Resources/2018_travelers.csv"

travel_2016_df = pd.read_csv(travel_2016)
travel_2017_df = pd.read_csv(travel_2017)
travel_2018_df = pd.read_csv(travel_2018)

# columns=['COUNTRY OF NATIONALITY']
# for x in travel_2018_df.columns:
#     if x !="COUNTRY OF NATIONALITY":
#         columns.append(f"{x}_2018")
travel_2016_df.columns = rename_columns(travel_2016_df.columns, "2016")
travel_2017_df.columns = rename_columns(travel_2017_df.columns, "2017")
travel_2018_df.columns = rename_columns(travel_2018_df.columns, "2018")
```
